```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math

def sign(x):
    return 1 if x >= 0 else -1

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

class Neuron:
    def __init__(self, input_size, lr=0.1):
        self.weights = np.random.uniform(-1, 1, size=input_size)
        self.lr = lr

    def predict(self, inputs):
        return np.sign(np.dot(inputs, self.weights))

    def fit(self, X, y, activation='sign', epochs=100):
        if activation == 'sign':
            activation_fn = sign
        elif activation == 'sigmoid':
            activation_fn = sigmoid

        for epoch in range(epochs):
            for i in range(X.shape[0]):
                inputs = X[i]
                target = y[i]
                output = activation_fn(np.dot(inputs, self.weights))
                error = target - output
                self.weights += self.lr * error * inputs

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.label = 1 if x >= y else -1

def generate_data(num_points):
    points = []
    for i in range(num_points):
        x = np.random.uniform(0, 700)
        y = np.random.uniform(0, 700)
        points.append(Point(x, y))
    return points

points = generate_data(100)

Nr = Neuron(input_size=2, lr=0.1)
print("Initial Weights:", Nr.weights)

for activation in ['sign', 'sigmoid']:
    Nr = Neuron(input_size=2, lr=0.1)
    print("Activation Function:", activation)
    for epoch in range(20):
        Nr.fit(np.array([[pt.x, pt.y] for pt in points]), np.array([pt.label for pt in points]), activation=activation, epochs=1)
        print("Weights after Epoch {}: {}".format(epoch+1, Nr.weights))
```

```python
import matplotlib.pyplot as plt
import numpy as np

activations=['sign','sigmoid']
for activation in activations:
    # Load the data points
    points = generate_data(100)

    # Train the neuron
    Nr = Neuron(input_size=2, lr=0.1)
    for i in range(20):
        for pt in points:
            inputs = [pt.x, pt.y]
            Nr.fit(np.array([inputs]), np.array([pt.label]), activation=activation, epochs=1)

    # Plot the data points
    colors = ['skyblue' if pt.label == -1 else 'orange' for pt in points]
    plt.scatter([pt.x for pt in points], [pt.y for pt in points], c=colors)

    # Plot the classification line
    x_vals = np.linspace(0, 700, 100)
    y_vals = [(-Nr.weights[0]*x - Nr.weights[1])/Nr.weights[1] for x in x_vals]
    if activation == 'sign':
        plt.plot(x_vals, y_vals, c='red', label='sign')
    elif activation == 'sigmoid':
        plt.plot(x_vals, sigmoid(y_vals), c='green', label='sigmoid')

plt.legend()
plt.show()
```

The generate_data function is used instead of loading data from a file.
The fit method is used instead of the undefined train method.
The inputs and target arguments are passed as numpy arrays to the fit method.
The equation for the classification line is calculated using the weights of the neuron, instead of calling an undefined estimate_y method.
The classification line is plotted using a red color for the 'sign' activation function and a green color for the 'sigmoid' activation function.
A legend is added to the plot to differentiate between the two activation functions.
